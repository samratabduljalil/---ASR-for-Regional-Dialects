```python
def pretty_sort(filename):
    name, number_str = filename.split(" (")
    number = int(number_str.split(")")[0])
    return name, number
```

```python
ids = []
preds=[]
```

```python
for root, dirs, files in os.walk("ben10/16_kHz_valid_audio"):
    files = sorted(files, key=pretty_sort)
    
#     print(files.index("valid_sandwip (1).wav"))
#     print(files.index("valid_sandwip (132).wav"))
    
#     put swandip first
    shift = files[1070 : 1202]
    
    files = shift + files[:1070] + files[1202:]
    ids = files.copy()
    
    for file in files:
        r=pretty_sort(file)
        if r[0] == "valid_sandwip":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_sandwip(audio)["text"]
            preds.append(text)
            print(text)
        if r[0] == "valid_barishal":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_barishal(audio)["text"]
            preds.append(text)
            print(text)
        if r[0] == "valid_chittagong":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_chittagong(audio)["text"]
            preds.append(text)
        if r[0] == "valid_habiganj":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_habiganj(audio)["text"]
            preds.append(text)
            print(text)
        if r[0] == "valid_kishoreganj":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_kishoreganj(audio)["text"]
            preds.append(text)
            print(text)
        if r[0] == "valid_narail":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_narail(audio)["text"]
            preds.append(text)
            print(text)
        if r[0] == "valid_narsingdi":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_narsingdi(audio)["text"]
            preds.append(text)
            print(text)
        if r[0] == "valid_rangpur":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_rangpur(audio)["text"]
            preds.append(text)
            print(text)
        if r[0] == "valid_sylhet":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_sylhet(audio)["text"]
            preds.append(text)
            print(text)
        if r[0] == "valid_tangail":
            print(r)
            composed_path = f"{test_data_dir}{file}"
            audio, sr = librosa.load(composed_path, sr=16_000)
            text = pipe_tangail(audio)["text"]
            preds.append(text)
            print(text)
```
